## Erros e Exceções

- Sintaxe está certa, mas variável não existe

````python
# não existe
print(x)
````

- Quando a sintaxe está certa, mas o comando não funciona, se chama exceção.
- **Exemplo**: Comando está certo, mas se eu digitar um valor diferente, ele retorna um ValueError.

````python
n = int(input('Número: '))
print(f'Você digitou {n}')
````

- Excessão numa operação (Dividir por Zero)

````python
a = int(input('Numerador: '))
b = int(input('Denominador: '))
r = a/b
print(f'O resultado é {r}')
````

## Tratamento de Erros

- `try:` Tenta executar um comando
- `except:` Caso ocorra uma exceção

```
try:
    {operação}
except:
    {falhou}
```
*Opcionais*

- `else:` caso dê certo
- `finally:` independente de erro/acerto

```
else:
    {deu certo}
finally:
    {certo/falha}
```
